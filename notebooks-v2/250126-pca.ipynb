```python
import numpy as np
import pandas as pd
import pathlib
import sklearn
```

```python
input = pathlib.Path('../datasets/numenta/monthly_sp500.csv')
output = pathlib.Path('../datasets/numenta/monthly_sp500_pca.csv')

df = pd.read_csv(input)
df.dropna(inplace=True)

col_date = df.pop('Date')

pca = sklearn.decomposition.PCA(n_components=1)
X_pca = pca.fit_transform(df.values)

df_pca = pd.DataFrame({'Date': col_date, 'pca': X_pca[:,0]})
df_pca.to_csv(output, index=False)

df_pca
```
